```python
import json
from firebase_admin import db

ref = db.reference("students")


# =========================
# NAME FORMAT
# =========================
def format_name(name):
    name = name.strip()

    if "," in name:
        parts = name.split(",")
        return parts[0].strip() + " " + parts[1].strip()

    parts = name.split()
    return parts[0] + " " + parts[1] if len(parts) >= 2 else name


# =========================
# SUBJECT UNITS
# =========================
UNITS = {
    "Mathematics": 1.7,
    "Computer Science": 1,
    "English": 1.3,
    "Filipino": 1,
    "Social Science": 1,
    "PEHM": 1,
    "Values Education": 0.7,
    "AdTech": 1,
    "Earth Science": 0.7
}


# =========================
# COMPUTE GWA
# =========================
def compute_gwa(subjects):
    total_units = 0
    total_weighted = 0

    for subject, grade in subjects.items():
        units = UNITS.get(subject, 1)
        total_weighted += grade * units
        total_units += units

    return total_weighted / total_units if total_units else 0


# =========================
# ADD STUDENT
# =========================
def add_student():
    name = format_name(input("\nEnter name (First, Last): "))
    section = input("Enter section: ")

    subjects = {}

    print("\nEnter subjects (type 'done'):")

    while True:
        subject = input("Subject: ")
        if subject.lower() == "done":
            break

        grade = float(input("Grade: "))
        subjects[subject] = grade

    failed = {s: g for s, g in subjects.items() if g < 75}
    gwa = compute_gwa(subjects)

    data = {
        "name": name,
        "section": section,
        "subjects": subjects,
        "failed_subjects": failed,
        "gwa": gwa
    }

    ref.child(name).set(data)

    print("\nAdded:", name)
    print("GWA:", round(gwa, 2))


# =========================
# SEARCH STUDENT
# =========================
def search_student():
    name = format_name(input("\nEnter name: "))
    data = ref.child(name).get()

    if not data:
        print("Not found")
        return

    print("\n==============================")
    print("PSHS-DRC SCHOOL DATABASE")
    print("==============================")
    print("Name:", data.get("name", "Unknown"))
    print("Section:", data.get("section", "Unknown"))
    print("GWA:", round(data.get("gwa", 0), 2))

    print("\nSubjects:")
    for s, g in data.get("subjects", {}).items():
        print(s + ":", g)


# =========================
# DELETE STUDENT
# =========================
def delete_student():
    name = format_name(input("\nEnter name: "))

    if ref.child(name).get():
        if input("Delete? (yes/no): ").lower() == "yes":
            ref.child(name).delete()
            print("Deleted")
    else:
        print("Not found")


# =========================
# ANALYTICS
# =========================
def get_all_students():
    return ref.get() or {}


def class_ranking():
    students = get_all_students()

    ranked = sorted(
        students.values(),
        key=lambda x: x.get("gwa", 0)
    )

    print("\nCLASS RANKING")

    for i, s in enumerate(ranked, 1):
        print(i, s.get("name", "Unknown"), "GWA:", round(s.get("gwa", 0), 2))


def section_average():
    section = input("Enter section: ")
    students = get_all_students()

    gwas = [
        s.get("gwa", 0)
        for s in students.values()
        if s.get("section") == section
    ]

    if not gwas:
        print("No data")
        return

    print("\nSection Average:", round(sum(gwas) / len(gwas), 2))


def full_report():
    students = get_all_students()

    print("\nFULL REPORT")

    for s in students.values():
        print("-------------------")
        print("Name:", s.get("name", "Unknown"))
        print("Section:", s.get("section", "Unknown"))
        print("GWA:", round(s.get("gwa", 0), 2))
        print("Subjects:", s.get("subjects", {}))


# =========================
# MAIN SYSTEM
# =========================
while True:

    print("\n==============================")
    print("PSHS-DRC SCHOOL DATABASE")
    print("==============================")

    print("""
1. Add Student
2. Search Student
3. Delete Student
4. Class Ranking
5. Section Average
6. Full Report
7. Exit
""")

    choice = input("Choose: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        search_student()

    elif choice == "3":
        delete_student()

    elif choice == "4":
        class_ranking()

    elif choice == "5":
        section_average()

    elif choice == "6":
        full_report()

    elif choice == "7":
        break

    else:
        print("Invalid input")
```
